Columns to drop

```python
main_label = 'accidently_injured'
raw_data[main_label] = (raw_data['NUMBER OF PERSONS INJURED']+raw_data['NUMBER OF PERSONS KILLED']>0).astype(int)

```

Handling null values 

```python
#fill null value
for col in [ 'BOROUGH', 'CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2', 
            'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']:
    raw_data[col] = raw_data[col].fillna('None')
raw_data.isnull().sum()
```

Extract year , hour from our crash date and date time columns

```python
# extract year
raw_data['Year'] = raw_data['CRASH DATE'].fillna('None').apply(lambda x: x[-4:])
# extract crash hour
raw_data['Hour'] = raw_data['CRASH TIME'].fillna('None').apply(lambda x: x.split(':')[0])
# extract day of week
raw_data['CRASH DATE'] = pd.to_datetime(raw_data['CRASH DATE'], format='%m/%d/%Y')
raw_data['Day_of_week'] = raw_data['CRASH DATE'].dt.day_name().astype(str)
```

Dropping coloumns which we wont use , also to minimize data to handle load

```python
cols2drop = ['ZIP CODE', 'LATITUDE', 'LONGITUDE', 'LOCATION', 'ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME', 'NUMBER OF PERSONS INJURED', 
             'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED', 
             'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED', 'CONTRIBUTING FACTOR VEHICLE 3',
             'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5', 'COLLISION_ID', 'VEHICLE TYPE CODE 3', 
             'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5', 'CRASH DATE', 'CRASH TIME']
raw_data = raw_data.drop(cols2drop, axis=1)
print(raw_data.shape)
```
